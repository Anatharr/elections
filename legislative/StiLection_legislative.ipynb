# Prepare data to be processed 

```python
def getDataFromDep(data, dep, tour=None):
    if tour is not None :
        data = data[data['num_tour']==tour]
    return data[data['code_departement'] == str(dep)]
    
    

def saveData(data, loc):
    # save cher_data as excel
    writer = pd.ExcelWriter(loc)
    
    # write dataframe to excel
    data.to_excel(writer)

    # save the excel
    writer.save()

def concatTarget(data, target_data):
    data = data.merge(target_data, how='left', on=['nom_commune', 'nom_candidat'] )
    return data
    
```

```python
cher_data             = getDataFromDep(df, 18, tour=1)
cher_data_t2          = getDataFromDep(df, 18, tour=2)
cher_data_test        = getDataFromDep(df_17, 18)
cher_data_test_result = getDataFromDep(df_17_t2, 18)


target_data_test = cher_data_test_result[['nom_commune', 'nom_candidat', 'voix']]
cher_data_test   = concatTarget(cher_data, target_data_test)
print(cher_data_test)

target_data = cher_data_t2[['nom_commune', 'nom_candidat', 'voix']]
cher_data = concatTarget(cher_data, target_data)


saveData(cher_data,'dataset/leg_12_cher.xlsx')
saveData(cher_data_t2, 'dataset/leg_12_cher_t2.xlsx')

cher_data.rename(columns={'voix_y':'voix_target', 'voix_x':'voix'}, inplace=True)
cher_data.voix_target.fillna(0, inplace=True)


```
